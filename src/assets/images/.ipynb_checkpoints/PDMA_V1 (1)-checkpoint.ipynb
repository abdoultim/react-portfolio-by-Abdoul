```python
# Importer les bibliothèques nécessaires
import os
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, to_timestamp, year, avg, max as spark_max
from pyspark.sql.window import Window
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.arima.model import ARIMA
import warnings
import folium
from branca.colormap import LinearColormap
from IPython.display import display, HTML
import logging
from pyspark.sql import Row
```

```python
# Ignorer les avertissements
warnings.filterwarnings('ignore')

# Configurer les variables d'environnement pour Spark
os.environ['SPARK_HOME'] = '/opt/spark'
os.environ['PYSPARK_PYTHON'] = 'python3'
os.environ['PYSPARK_DRIVER_PYTHON'] = 'python3'
```

```python
# Réduire le niveau de journalisation de Spark pour ignorer les messages WARN
logger = logging.getLogger('py4j')
logger.setLevel(logging.ERROR)
# Créer une session Spark
spark = SparkSession.builder.appName("TimeSeriesAnalysis").getOrCreate()
```

# 2- Operations avec spark

## 2-1 Création des dataframes

```python
# Créer un DataFrame à partir d'une liste de tuples
data = [("Alice", 34), ("Bob", 45), ("Cathy", 29)]
columns = ["Nom", "Age"]
df = spark.createDataFrame(data, columns)

# Afficher le contenu du DataFrame
df.show()
```

```python
# Créer un DataFrame Spark avec des données factices
data = [
    Row(date="2022-01-01", value=50),
    Row(date="2022-01-02", value=65),
    Row(date="2021-01-03", value=90),
    Row(date="2021-01-04", value=11),
    Row(date="2020-01-05", value=39),
    Row(date="2020-01-03", value=36),
    Row(date="2019-01-04", value=70),
    Row(date="2019-01-03", value=40),
    Row(date="2019-01-04", value=25),
]

df = spark.createDataFrame(data)
df.show()
```

## 2-2 Manipulation des données

```python
# Convertir la colonne 'date' en type timestamp
df = df.withColumn("date", to_timestamp(col("date"), "yyyy-MM-dd"))
df.show()
```

```python
# Extraire l'année et calculer la moyenne des valeurs par année
df = df.withColumn("year", year(col("date")))
df_avg = df.groupBy("year").agg(avg("value").alias("avg_value"))

# Trier les années dans l'ordre croissant
df_avg = df_avg.orderBy("year")

df_avg.show()
```

## 2-3 Visualisation avec Matplotlib

```python
# Collecter les données pour la visualisation
data = df_avg.collect()
years = [row['year'] for row in data]
avg_values = [row['avg_value'] for row in data]

# Créer un graphique
plt.figure(figsize=(10, 5))
plt.plot(years, avg_values, marker='o')
plt.title("Average Value per Year")
plt.xlabel("Year")
plt.ylabel("Average Value")
plt.grid(True)
plt.show()
```

## SERIES TEMPORELLES AVEC STATSMODELS

## 1- Chargements des bases de données

```python
# Lire les fichiers CSV
df = spark.read.csv('PDMA/france.csv.gz', header=True, inferSchema=True)
stations = spark.read.csv("PDMA/postesSynop.csv", sep=";", header=True, inferSchema=True)
```

```python
df.show()
```

```python
# Préparer les données
df = df.withColumn("date", to_timestamp(col("date")))
df = df.withColumn("année", year(col("date")))
df
```

```python
# Filtrer les stations en métropole
stations_filtered = stations.filter((col("Latitude") > 40) & (col("Longitude") > -10))
metropole_ids = [row['ID'] for row in stations_filtered.select("ID").collect()]
df = df.filter(col("numer_sta").isin(metropole_ids))
```

```python
# Calculer la température maximale par station
window_spec = Window.partitionBy("numer_sta")
max_temp_df = df.withColumn("tmax", spark_max("t").over(window_spec))
max_temp_df = max_temp_df.groupBy("numer_sta").agg(spark_max("tmax").alias("tmax"), spark_max("date").alias("datemax"))
```

```python
# Joindre avec les informations des stations
result_df = max_temp_df.join(stations, max_temp_df["numer_sta"] == stations["ID"]).select("numer_sta", "tmax", "datemax", "Latitude", "Longitude", "Nom")
```

```python
# Convertir la température de Kelvin à Celsius
result_df = result_df.withColumn("tmax_celsius", col("tmax") - 273.15)
```

```python
# Créer la carte avec Folium
result_pd = result_df.toPandas()
carte = folium.Map(location=[46.7, 0.6], zoom_start=6)
colormap = LinearColormap(colors=['blue', 'green', 'yellow', 'red'], vmin=result_pd['tmax_celsius'].min(), vmax=result_pd['tmax_celsius'].max())
colormap.caption = 'Température (°C)'

for idx, row in result_pd.iterrows():
    color = colormap(row['tmax_celsius'])
    folium.CircleMarker(
        location=[row['Latitude'], row['Longitude']],
        radius=5,
        color=color,
        fill=True,
        fill_color=color,
        fill_opacity=0.7,
        tooltip=f'{row.tmax_celsius:.2f} °C<br/>{row.Nom}<br/>{row.datemax}'
    ).add_to(carte)

colormap.add_to(carte)
carte.save("carte_temperature_maximale.html")
display(HTML('<iframe src="carte_temperature_maximale.html" width="100%" height="600"></iframe>'))
```

```python
# Calculer la température moyenne annuelle
annual_avg_df = df.groupBy("année").agg(avg("t").alias("temp_moyenne"))
annual_avg_df = annual_avg_df.filter(col("année") <= 2022)
annual_avg_df = annual_avg_df.withColumn("temp_moyenne_celsius", col("temp_moyenne") - 273.15)
```

```python
# Convertir les résultats en Pandas DataFrame pour l'analyse
annual_avg_pd = annual_avg_df.toPandas()
annual_avg_pd = annual_avg_pd.sort_values(by='année')
annual_avg_pd.reset_index(drop=True, inplace=True)
# Mettre la colonne 'année' en index
annual_avg_pd.set_index('année', inplace=True)
```

```python
annual_avg_pd
```

```python
# Visualisation des données en Celsius
plt.figure(figsize=(10, 6))
plt.plot(annual_avg_pd['temp_moyenne_celsius'], label='Température moyenne annuelle')
plt.xlabel('Année')
plt.ylabel('Température (°C)')
plt.title('Température moyenne annuelle en France')
plt.legend()
plt.show()
```

```python
# Analyse ACF et PACF
plt.figure(figsize=(12, 6))
plt.subplot(121)
plot_acf(annual_avg_pd['temp_moyenne_celsius'], lags=12, ax=plt.gca())
plt.subplot(122)
plot_pacf(annual_avg_pd['temp_moyenne_celsius'], lags=12, ax=plt.gca())
plt.show()
```

```python
# Séparer les données en ensemble d'entraînement et de test
train_data = annual_avg_pd['temp_moyenne_celsius'][:-22]
test_data = annual_avg_pd['temp_moyenne_celsius'][-22:]
```

```python
# Créer et ajuster le modèle ARIMA
model = ARIMA(train_data, order=(1, 1, 1))
model_fit = model.fit()

# Afficher le résumé du modèle
print(model_fit.summary())
```

```python
# Ajustement du modèle ARIMA
model = ARIMA(annual_avg_pd['temp_moyenne_celsius'], order=(1, 1, 1))  # (p, d, q) paramètres de l'ARIMA
model_fit = model.fit()

forecast_steps = 5  # Prévisions pour les 5 prochaines années
forecast = model_fit.forecast(steps=forecast_steps)

# Créer une liste des années de prévision
forecast_years = list(range(2023, 2023 + len(forecast)))
```

```python
# Visualisation des prévisions
plt.figure(figsize=(10, 6))
plt.plot(annual_avg_pd.index, annual_avg_pd['temp_moyenne_celsius'], marker='o', label='Données réelles')
plt.plot(forecast_years, forecast, marker='o', linestyle='--', color='red', label='Prévisions')
plt.title('Prévisions des températures moyennes annuelles')
plt.xlabel('Année')
plt.ylabel('Température moyenne (°C)')
plt.legend()
plt.grid(True)
plt.show()
```
