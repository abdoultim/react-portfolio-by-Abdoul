```python
import pandas as pd
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
```

```python
import pandas as pd
import folium
from branca.colormap import LinearColormap
from IPython.display import display, HTML

print("carte_temperature_maximale.html")
print("Carte illustrant la temperature maximale en °C par station depuis 1996 et la date d'enregistrement")

# Lire les fichiers CSV
T = pd.read_csv('france.csv.gz', parse_dates=['date'])
stations = pd.read_csv("postesSynop.csv", sep=";")

# Préparer les données
t = T.copy()  # Créez une copie du DataFrame T pour éviter de modifier l'original
t['ID'] = t['numer_sta']

result = t.groupby('numer_sta').idxmax()
y = pd.DataFrame(t['date'].iloc[result['t']])
y.index = result.index

x = pd.DataFrame(t['t'].iloc[result['t']])
x.index = result.index

result['datemax'] = y['date']
result['tmax'] = x['t']
result['ID'] = result.index
df = pd.merge(stations, result, on='ID', how='inner')

df = df[df['tmax'] < 330]
df['tmax'] = df['tmax'] - 273.15

# Créer la carte
carte = folium.Map(location=[46.7, 0.6], zoom_start=6)

# Créer une échelle de couleur basée sur la colonne "tmax"
colormap = LinearColormap(colors=['blue', 'green', 'yellow', 'red'], vmin=df['tmax'].min(), vmax=df['tmax'].max())
colormap.caption = 'Température (°C)'

for idx, row in df.iterrows():
    # Obtenir la couleur de l'échelle en fonction de la valeur de "tmax"
    color = colormap(row['tmax'])

    # Créer un cercle (CircleMarker) avec la couleur correspondante
    circle_marker = folium.CircleMarker(
        location=[row['Latitude'], row['Longitude']],
        radius=5,
        color=color,
        fill=True,
        fill_color=color,
        fill_opacity=0.7,
        tooltip=f'{row.tmax:.2f} °C<br/>{row.Nom}<br/>{row.datemax}'
    )
    circle_marker.add_to(carte)

# Ajouter l'échelle de couleur à la carte
colormap.add_to(carte)

# Sauvegarder la carte
carte.save("carte_temperature_maximale.html")

# Afficher la carte dans Jupyter Notebook
display(HTML('<iframe src="carte_temperature_maximale.html" width="100%" height="600"></iframe>'))
```

```python
stations = pd.read_csv("postesSynop.csv", sep=";")
metropole = stations.ID[(stations.Latitude > 40) & (stations.Longitude > -10)]

t = pd.read_csv('france.csv.gz', parse_dates=['date'])
t['année'] = t.date.apply(lambda x: x.year)    
t = t[t.numer_sta.isin(metropole) & (t.année <= 2022)]
moy = t.groupby('année').mean(['t']).reset_index()

def tendance(depuis=1996, jusqua=2022):
    selection = moy[(moy.année >= depuis) & (moy.année <= jusqua)]
    lin = smf.ols('t ~ année', data=selection).fit()
    tendance = 'Tendance %d-%d : %.2f °C/décade' % (
        depuis, jusqua, 10*lin.params.année)
    plt.plot(selection.année, lin.fittedvalues, label=tendance)

plt.plot(moy.année, moy.t, label='Moyenne annuelle')
tendance()
tendance(2010)

plt.title('France')
plt.xlabel('Année')
plt.ylabel('Température (K)')
plt.legend()
plt.savefig('France.pdf')
```
