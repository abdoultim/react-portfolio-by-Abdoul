```python
import pandas as pd
import numpy as np
import requests
import bs4
import re
import matplotlib as plt
import plotly.graph_objs as go
import matplotlib.pyplot as pltd
from statsmodels.tsa.vector_ar.var_model import VAR
from statsmodels.tsa.stattools import adfuller
from statsmodels.tools.eval_measures import rmse, aic
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python

web_page_url = "https://www.politologue.com/prix-carburants/"
web_page_reponse = requests.get(web_page_url)
soup = bs4.BeautifulSoup(web_page_reponse.content, 'html.parser')

# Récupérer les prix
prix = soup.find_all(name="td", class_="bold")  # Limite à 122 premiers prix
montants = []
for p in prix:
    valeur = p.text.strip()
    montant = re.search(r'([\d,]+)\D', valeur)  # Extraire le montant
    if montant:
        montants.append(float(montant.group(1).replace(',', '.')))


colonnes_prix = ['Gasoil', 'SP98', 'SP95', 'E10', 'E85', 'GPL']
data_prix = [montants[i:i+6] for i in range(0, len(montants), 6)]  # Diviser les montants en listes de 6 pour chaque type de carburant
df_prix = pd.DataFrame(data_prix, columns=colonnes_prix)  # Créer le DataFrame pour les prix

# Récupérer les Mois
Mois = soup.find_all(name="td", class_="")[:88]  # Limite à 124 premières Mois
liste_Mois = [d.text.strip() for d in Mois]  # Extraire les Mois
df_Mois = pd.DataFrame({'Mois': liste_Mois})  # Créer le DataFrame pour les Mois
df_Mois['Mois'] = df_Mois['Mois'].replace({'Avril - 2024*': 'Avril - 2024'})


# Afficher les DataFrames
fuel = pd.merge(df_Mois, df_prix, left_index=True, right_index=True, how='inner')
fuel = fuel.iloc[::-1]
fuel.to_csv('fuel.csv', index=False)
```

```python
import pandas as pd

# Supposons que fuel est votre DataFrame et que la colonne "Mois" contient vos données
try:
    # Convertir la colonne "Mois" en objets datetime
    pd.to_datetime(fuel['Mois'])
    print("La colonne 'Mois' contient des dates.")
except ValueError:
    print("La colonne 'Mois' ne contient pas des dates.")
```

```python
def format_date(mois):
    mois_str, annee = mois.split(" - ")
    mois_numerique = {
        'Janvier': '01',
        'Février': '02',
        'Mars': '03',
        'Avril': '04',
        'Mai': '05',
        'Juin': '06',
        'Juillet': '07',
        'Août': '08',
        'Septembre': '09',
        'Octobre': '10',
        'Novembre': '11',
        'Décembre': '12'
    }
    mois_num = mois_numerique.get(mois_str)
    if mois_num is None:
        raise ValueError("Mois invalide")
    # Notre format souhaité (JJ/MM/AAAA)
    return f"01/{mois_num}/{annee}"

fuel['Mois'] = fuel['Mois'].apply(format_date)
print(fuel['Mois'])
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt

# Supposons que fuel est votre DataFrame contenant les données sur les prix du carburant

# Création d'un modèle de régression linéaire pour chaque type de carburant
models = {}

# Boucle sur tous les types de carburant à l'exception de la variable 'Mois'
for carburant in fuel.columns[1:]:
    # Extrait des données pour la régression linéaire
    X = np.arange(len(fuel)).reshape(-1, 1)  # Numéro de l'observation comme variable explicative
    y = fuel[carburant]  # Prix du carburant comme variable cible

    # Création d'un modèle de régression linéaire
    model = LinearRegression()

    # Ajustement du modèle aux données
    model.fit(X, y)

    # Enregistrer le modèle dans le dictionnaire avec le nom du carburant comme clé
    models[carburant] = model

    # Prédiction des valeurs avec le modèle ajusté
    y_pred = model.predict(X)

    # Visualisation des résultats
    plt.figure(figsize=(20, 12))
    plt.plot(fuel['Mois'], y, label='Données réelles')
    plt.plot(fuel['Mois'], y_pred, label='Régression linéaire', linestyle='--')
    plt.title(f'Régression linéaire pour les prix du carburant {carburant}')
    plt.xlabel('Mois et année')
    plt.ylabel('Prix du carburant (en €)')
    plt.xticks(rotation=45)
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # Coefficients de régression
    print(f"Pour le carburant {carburant}:")
    print("Coefficient de pente (tendance) :", model.coef_[0])
    print("Coefficient d'interception :", model.intercept_)
```

# Analyse des Series Temporelles

Nous allons utilisé le modèle VAR (Vector Auto Regression) pour capturer les relations linéaires entre les differentes séries temporelles.

```python
fuel1 = fuel.loc[:, ['Mois', 'SP98]]
```

```python
#fuel1 = fuel.loc[:, ['Mois', 'SP98', 'SP95']]
# Mettre la date en index de la table
fuel1.set_index('Mois',inplace=True)
fuel1
```

```python
# Description des données
# Créer une figure et des axes
fig, axs = plt.subplots(3, 1, figsize=(10,15))

# Faire le premier sous-graphique pour SP98
axs[0].plot(fuel1.index, fuel1['SP98'], color='blue')
axs[0].set_title('Évolution du prix du SP98')

# Faire le deuxième sous-graphique pour SP95
axs[1].plot(fuel1.index, fuel1['E10'], color='red')
axs[1].set_title('Évolution du prix du SP95')

# Faire le troisième sous-graphique pour SP98 et SP95
axs[2].plot(fuel1.index, fuel1['SP98'], color='blue', label='SP98')
axs[2].plot(fuel1.index, fuel1['E10'], color='red', label='SP95')
axs[2].set_title('Évolution conjointe du prix du SP98 et SP95')
axs[2].legend()

# Afficher le graphique
plt.tight_layout()
plt.show()
```

```python
# Faire les tests de Dickey-Fuller augmenté
for column in ['SP98', 'E10']:
    result = adfuller(fuel1[column])
    print(f"\nResults for column {column}:")
    print('Adfuel1 Statistic: %f' % result[0])
    print('p-value: %f' % result[1])
    print('Critical Values:')
    for key, value in result[4].items():
        print('\t%s: %.3f' % (key, value))

# Faire les graphiques ACF et PACF pour chaque colonne
for column in ['SP98', 'E10']:
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(16, 8))
    plot_acf(fuel1[column], lags=min(len(fuel1[column])-1, 43), zero=False, ax=ax1)
    ax1.set_title(f"ACF de {column}")
    plot_pacf(fuel1[column], lags=min(len(fuel1[column])-1, 43), zero=False, ax=ax2)
    ax2.set_title(f"PACF {column}")
    plt.tight_layout()
    plt.show()
```

```python
# Différencier les séries de données et les ajouter au DataFrame pour rendre la serie stationnaire
fuel1_diff = fuel1.diff().dropna()
fuel1_diff.columns = [f"{col}_diff" for col in fuel1.columns]
fuel1 = pd.concat([fuel1, fuel1_diff], axis=1)
fuel1
```

```python
fuel1=fuel1.dropna()
fuel1
```

```python
# Tracer les graphiques
plt.figure(figsize=(10, 6))
plt.subplot(2, 1, 1)
plt.plot(fuel1_diff.index, fuel1_diff['SP98_diff'])
plt.title('Évolution de la série différenciée "SP98"')
plt.ylabel('Différence')

plt.subplot(2, 1, 2)
plt.plot(fuel1_diff.index, fuel1_diff['E10_diff'])
plt.title('Évolution de la série différenciée "SP95"')
plt.xlabel('Date')
plt.ylabel('Différence')

plt.tight_layout()
plt.show()
```

```python
# Faire les tests de Dickey-Fuller augmenté pour les séries différenciées
for column in fuel1_diff.columns:
    result = adfuller(fuel1_diff[column])
    print(f"\nResults for column {column}:")
    print('Adfuel1 Statistic: %f' % result[0])
    print('p-value: %f' % result[1])
    print('Critical Values:')
    for key, value in result[4].items():
        print('\t%s: %.3f' % (key, value))

# Faire les graphiques ACF et PACF pour chaque colonne différenciée
for column in fuel1_diff.columns:
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(16, 8))
    plot_acf(fuel1_diff[column], lags=10, zero=False, ax=ax1)
    ax1.set_title(f"ACF de {column}")
    plot_pacf(fuel1_diff[column], lags=10, zero=False, ax=ax2)
    ax2.set_title(f"PACF de {column}")
    plt.tight_layout()
    plt.show()
```

```python
# Définir le point de division
split_point = int(len(fuel1_diff) * 0.9)  # 90% des données pour l'entraînement

# Diviser les données
fuel1_train = fuel1_diff.iloc[:split_point]
fuel1_test = fuel1_diff.iloc[split_point:]

# Vérifier les tailles des ensembles d'entraînement et de test
print('Training set:', len(fuel1_train))
print('Test set:', len(fuel1_test))
```

```python
# Choisir l'ordre optimal du modèle VAR
model = VAR(fuel1_train)

# Choisir l'ordre optimal p pour le modèle VAR
order_selection = model.select_order(maxlags=6)
order_selection.summary()
```

```python

# L'ordre optimal est donné par le critère d'information d'Akaike (AIC) ou le critère d'information bayésien (BIC) 
optimal_order_aic = order_selection.aic
optimal_order_bic = order_selection.bic
optimal_order_fpe = order_selection.fpe
optimal_order_hqic = order_selection.hqic

print(f"P optimal avec AIC: {optimal_order_aic}")
print(f"P optimal avec BIC: {optimal_order_bic}")
print(f"P optimal avec FPE: {optimal_order_fpe}")
print(f"P optimal avec HQIC: {optimal_order_hqic}")
```

```python
# Créer le modèle VAR
model = VAR(fuel1_train)

# Choisir un ordre 5
results = model.fit(3)

# Voir un sommaire des résultats
print(results.summary())
```

```python
# Obtenir les résidus
residuals = results.resid
residuals
```

```python
# Tracer l'ACF et le PACF pour chaque série de résidus
for col in residuals.columns:
    fig, axes = plt.subplots(1, 2, figsize=(15, 5))
    
    plot_acf(residuals[col], ax=axes[0])
    axes[0].set_title(f'Autocorrélation des résidus pour {col}')
    
    plot_pacf(residuals[col], ax=axes[1])
    axes[1].set_title(f'Autocorrélation partielle des résidus pour {col}')

plt.show()
```

```python
from statsmodels.stats.stattools import jarque_bera
from statsmodels.stats.diagnostic import acorr_ljungbox
# Effectuer des tests sur les résidus
for col in residuals.columns:
    print(f"Résidus pour {col}:")
    jb_test = jarque_bera(residuals[col])
    print(f"Test de Jarque-Bera: statistic={jb_test[0]}, p-value={jb_test[1]}")
```

# Prediction train  et test

```python
# Préparer un tableau pour stocker les résultats
results_df = pd.DataFrame(columns=['Variable', 'Lag', 'LB Statistic', 'LB P-value'])
```

```python
# Préparer un tableau pour stocker les résultats
results_list = []

# Effectuer le test de Ljung-Box pour des décalages de 1 à 12
for col in residuals.columns:
    for lag in range(1, 13):
        lb_test = acorr_ljungbox(residuals[col], lags=[lag], return_df=True)
        results_list.append({
            'Variable': col,
            'Lag': lag,
            'LB Statistic': lb_test['lb_stat'].values[0],
            'LB P-value': lb_test['lb_pvalue'].values[0]
        })

# Convertir la liste de dictionnaires en DataFrame
results_df = pd.DataFrame(results_list)

# Afficher les résultats
print(results_df)
```

```python
# Prévoir les valeurs sur l'ensemble d'entraînement
# Prédictions sur les ensembles d'entraînement et de test
train_pred = results.fittedvalues
test_pred = results.forecast(fuel1_train.values, steps=len(fuel1_test))
```

```python
# Créer un DataFrame avec les prédictions
train_pred_df = pd.DataFrame(train_pred, columns=fuel1_train.columns)
train_pred_df['Mois'] = fuel1_diff.index[len(fuel1_train)]  # Ajouter les dates correspondant à l'ensemble d'entraînement
train_pred_df.set_index('Mois', inplace=True)  # Utiliser les dates comme indices

test_pred_df = pd.DataFrame(test_pred, columns=fuel1_train.columns)
test_pred_df['Mois'] = fuel1_diff.index[len(fuel1_train):]  # Ajouter les dates correspondant à l'ensemble d'entraînement
test_pred_df.set_index('Mois', inplace=True)  # Utiliser les dates comme indices
```
