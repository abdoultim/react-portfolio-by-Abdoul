#Description des variables 

USA State : Il sagit du nom de lÉtat des États-Unis concerné par les données.

Total Cases : Cest le nombre total de cas confirmés de COVID-19 dans cet État depuis le début de la pandémie.

Total Deaths : Il sagit du nombre total de décès attribués au COVID-19 dans cet État depuis le début de la pandémie.

Total Recovered : Cest le nombre total de personnes considérées comme récupérées du COVID-19 dans cet État depuis le début de la pandémie.

Active Cases : Cest le nombre total de cas actifs de COVID-19 dans cet État à un moment donné, cest-à-dire le nombre total de cas moins les récupérations et les décès.

Tot Cases/ 1M pop : Cest le nombre total de cas de COVID-19 pour chaque million de personnes dans la population de cet État. Cela permet de normaliser les cas par rapport à la taille de la population.

Deaths/ 1M pop : Cest le nombre total de décès attribués au COVID-19 pour chaque million de personnes dans la population de cet État. Encore une fois, cela permet de normaliser les décès par rapport à la taille de la population.

Total Tests : Cest le nombre total de tests de dépistage du COVID-19 réalisés dans cet État depuis le début de la pandémie.

Tests/ 1M pop : Cest le nombre total de tests de dépistage du COVID-19 réalisés pour chaque million de personnes dans la population de cet État. Encore une fois, cela permet de normaliser les tests par rapport à la taille de la population.

Population : Il sagit de la population totale de cet État.

```python
# Calculer la moyenne de chaque colonne
mean_total_recovered = data['Total Recovered'].mean()
mean_active_cases = data['Active Cases'].mean()

# Imputer les valeurs manquantes par la moyenne de chaque colonne
data['Total Recovered'].fillna(mean_total_recovered, inplace=True)
data['Active Cases'].fillna(mean_active_cases, inplace=True)
# Définir le format d'affichage des nombres à virgule flottante
pd.options.display.float_format = '{:.1f}'.format
```

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

url = "https://developers.google.com/public-data/docs/canonical/states_csv"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")

# Trouver le tableau contenant les données des États
table = soup.find("table")

# Initialiser des listes pour stocker les données
us_state_names = []
latitudes = []
longitudes = []
state_codes = []

# Parcourir les lignes du tableau
for row in table.find_all("tr")[1:]:  # Ignorer la première ligne (en-têtes)
    columns = row.find_all("td")
    if len(columns) >= 4:  # S'assurer que la ligne contient les informations nécessaires
        state_code = columns[0].text.strip()
        latitude = columns[1].text.strip()
        longitude = columns[2].text.strip()
        state_name = columns[3].text.strip()
        state_codes.append(state_code)
        latitudes.append(latitude)
        longitudes.append(longitude)
        us_state_names.append(state_name)

# Créer un DataFrame Pandas avec les données
data = {
    'State Code': state_codes,
    'USA State': us_state_names,
    'Latitude': latitudes,
    'Longitude': longitudes
}
df = pd.DataFrame(data)
df['Latitude'] = df['Latitude'].str.replace(',', '.').astype(float).round(6)
df['Longitude'] = df['Longitude'].str.replace(',', '.').astype(float).round(6)

# Écrire le DataFrame dans un fichier CSV
df.to_csv('usa_data.csv', index=False)
```
